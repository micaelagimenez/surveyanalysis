## Sorting the data into right and wrong answers

```python
#take data and decide which are right and which are not

#1
jerarquia_texto = data['jerarquia_texto']
def jerarquiatexto(jerarquia_texto):
    if jerarquia_texto == 'La capacidad general de procesar/estimar cantidades y la habilidad matemática dependiente de la instrucción.': return 'Correcto'
    else: return 'Incorrecto'

data['jerarquiatexto'] = data['jerarquia_texto'].map(jerarquiatexto)

#2
cohesion_estrucsintac = data['cohesion_estructurasintactica']
def cohesionsintactica(cohesion_estrucsintac):
    if cohesion_estrucsintac == 'Recién cuando maneja la noción de número, el niño es capaz de llevar a cabo operaciones de seriación y clasificación jerárquica.': return 'Correcto'
    else: return 'Incorrecto'

data['cohesionestructurasintactica'] = data['cohesion_estructurasintactica'].map(cohesionsintactica)

#3
inferencia = data['inferencia']
def inferencias(inferencia):
    if inferencia == 'El cerebro de los niños no puede incorporar realmente esa clase de conceptos antes de esa edad por una cuestión madurativa.': return 'Correcto'
    else: return 'Incorrecto'

data['inferencia1'] = data['inferencia'].map(inferencias)

#4
jerarquia_texto2 = data['jerarquia_texto2']
def jerarquiatexto2(jerarquia_texto2):
    if jerarquia_texto2 == 'Se consideró que eran limitadas por cuestiones metodológicas.': return 'Correcto'
    else: return 'Incorrecto'

data['jerarquia_texto2'] = data['jerarquia_texto2'].map(jerarquiatexto2)

#5
inferencia2 = data['inferencia2']
def inferencia2(inferencia2):
    if inferencia2 == 'Porque consideraban que el aspecto motivacional podía estar influyendo en el resultado.': return 'Correcto'
    else: return 'Incorrecto'

data['inferencia2'] = data['inferencia2'].map(inferencia2)

#6
semanticalexica = data['semanticalexica']
def semanticalexica(semanticalexica):
    if semanticalexica == 'De manera oculta.': return 'Correcto'
    else: return 'Incorrecto'

data['semanticalexica'] = data['semanticalexica'].map(semanticalexica)

#7
inferencia3 = data['inferencia3']
def inferencia3(inferencia3):
    if inferencia3 == 'El uso implícito de las estimaciones de cantidad.': return 'Correcto'
    else: return 'Incorrecto'

data['inferencia3'] = data['inferencia3'].map(inferencia3)

#8
modelomental = data['modelomental']
def modelomental(modelomental):
    if modelomental == 'Sorprendidos.': return 'Correcto'
    else: return 'Incorrecto'

data['modelomental'] = data['modelomental'].map(modelomental)

#9
semanticalexica_inferencia = data['semanticalexica_inferencia']
def semanticalexica_inferencia(semanticalexica_inferencia):
    if semanticalexica_inferencia == 'Que desde el nacimiento una parte del cerebro rige puntualmente la capacidad de procesar números y no otro tipo de información.': return 'Correcto'
    else: return 'Incorrecto'

data['semanticalexica_inferencia'] = data['semanticalexica_inferencia'].map(semanticalexica_inferencia)

#10
cohesion = data['cohesion']
def cohesion(cohesion):
    if cohesion == 'Que los niños de tres años estiman cantidades no por abstracción conceptual sino por la posición y forma de los objetos.': return 'Correcto'
    else: return 'Incorrecto'

data['cohesion'] = data['cohesion'].map(cohesion)

#11
intuiciontexto = data['intuiciontexto']
def intuiciontexto(intuiciontexto):
    if intuiciontexto == 'En años que siguieron, se realizaron varias pruebas que mostraron que los niños, desde muy temprana edad, tienen un cierto conocimiento de la cantidad.': return 'Correcto'
    else: return 'Incorrecto'

data['intuiciontexto'] = data['intuiciontexto'].map(intuiciontexto)

#12
flexibilidadcognitiva = data['flexibilidadcognitiva']
def flexibilidadcognitiva(flexibilidadcognitiva):
    if flexibilidadcognitiva == 'Reparar en el tipo de vocabulario oportuno para cada tipo textual.': return 'Correcto'
    else: return 'Incorrecto'

data['flexibilidadcognitiva'] = data['flexibilidadcognitiva'].map(flexibilidadcognitiva)

#13
cohesion2 = data['cohesion2']
def cohesion2(cohesion2):
    if cohesion2 == 'La diferencia entre la capacidad de manejar una aritmética básica y realizar operaciones matemáticas de mayor complejidad.': return 'Correcto'
    else: return 'Incorrecto'

data['cohesion2-'] = data['cohesion2'].map(cohesion2)

#14
hechos_consecucencias = data['hechos_consecucencias']
def hechos_consecucencias(hechos_consecucencias):
    if hechos_consecucencias == '(1) Los niños estiman cantidades; (2) emerge el lenguaje; (3) se desarrollan las habilidades matemáticas.': return 'Correcto'
    else: return 'Incorrecto'

data['hechos_consecucencias'] = data['hechos_consecucencias'].map(hechos_consecucencias)

#15
estructurasintactica = data['estructurasintactica']
def estructurasintactica(estructurasintactica):
    if estructurasintactica == 'Los niños poseen capacidades innatas. Sin embargo, las habilidades matemáticas son dependientes de la instrucción.': return 'Correcto'
    else: return 'Incorrecto'

data['estructurasintactica'] = data['estructurasintactica'].map(estructurasintactica)

data.head()
```

## Dividing data by age group

```python
#divide data by age group
menores, mayores = [x for _, x in data.groupby(data['Edad:'] > 30)]
```

## Create dataframes by age group with the needed columns

```python
#create menores df
menores_df = menores.drop(["Edad:","Género","Estudio",'Estudio_actual', 'Porcentaje_carrera', 'Carrera', 'Institucion', 'Carrera completa', 'Institucionterminada', 'horas_lectura',
              "lectura_pantalla","libros_anuales","momento_lectura","Hora lectura de texto","Código:"],axis=1)

menores_df['cohesion_estructurasintactica'] = menores_df['cohesion_estructurasintactica']
```

```python
#create mayores df
mayores_df = mayores.drop(["Edad:","Género","Estudio",'Estudio_actual', 'Porcentaje_carrera', 'Carrera', 'Institucion', 'Carrera completa', 'Institucionterminada', 'horas_lectura',
              "lectura_pantalla","libros_anuales","momento_lectura","Hora lectura de texto","Código:"],axis=1)
```

## Melt dataframes to get the right and wrong counts by electronics used

```python
#melt things into one dataframe> Mayores
dfs = mayores_df.set_index('dispositivo_encuesta')
right_counts = (dfs == "Correcto").astype('int').groupby('dispositivo_encuesta').sum()
wrong_counts = (dfs == "Incorrecto").astype('int').groupby('dispositivo_encuesta').sum()
print(wrong_counts)
print(right_counts)
```

```python
#melt things into one dataframe > Menores
df2 = menores_df.set_index('dispositivo_encuesta')
right_counts2 = (df2 == "Correcto").astype('int').groupby('dispositivo_encuesta').sum()
wrong_counts2 = (df2 == "Incorrecto").astype('int').groupby('dispositivo_encuesta').sum()
print(wrong_counts2)
print(right_counts2)
```

## Get the means by electronics used and then the general mean

```python
#get the menores means
mean_menores_wrong = wrong_counts2.mean(axis=1)
print("Menores wrong answers mean is:", 
      mean_menores_wrong)

mean_menores_right = right_counts2.mean(axis=1)
print("Menores right answers mean is:", 
      mean_menores_right)
```

```python
#mayores means
mean_mayores_wrong = wrong_counts.mean(axis=1)
print("Mayores wrong answers mean is:",
      mean_mayores_wrong)

mean_mayores_right = right_counts.mean(axis=1)
print("Mayores right asnwers mean is:",
      mean_mayores_right)
```

## Get the standard deviation

```python
#standard deviation 
stdev_mayores_right = np.std(right_counts, axis=1)
print("Mayores right answers standard deviation is:", stdev_mayores_right)

stdev_mayores_wrong = np.std(wrong_counts, axis=1)
print("Mayores wrong answers standard deviation is:", stdev_mayores_wrong)

stdev_menores_right = np.std(right_counts2, axis=1)
print("Menores right answers standard deviation is:", stdev_menores_right)

stdev_menores_wrong = np.std(wrong_counts2, axis=1)
print("Menores wrong answers standard deviation is:", stdev_menores_wrong)
```
